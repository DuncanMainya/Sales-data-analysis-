Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading Data

```python
data = pd.read_csv(r'C:\Users\Dan\Desktop\sales_analysis\data\cleaned_superstore.csv')
df = pd.DataFrame(data) #Creating a DataFrame
```

Descriptive Statistics

Total Sales

```python
total_sales = df['sales'].sum().round(2)
print(f"Total Sales: ${total_sales:,}")
```

Average Sales

```python
#Turning to datetime dtype
df['ship_date'] = pd.to_datetime(df['ship_date'],errors='coerce')

#Extracting Year, Month and Day
df['Year'] = df['ship_date'].dt.year
df['Month'] = df['ship_date'].dt.month
df['Day'] = df['ship_date'].dt.day

#Average sales per year
avg_sales_year = df.groupby(['Year'])['sales'].mean().round(2)
print("Average Sales Per Year:")
print(avg_sales_year)

#Average sales per month
avg_sales_month = df.groupby(['Month'])['sales'].mean().round(2)
print("\nAverage Sales Per Month:")
print(avg_sales_month)

#Average Sales Per day
avg_sales_day = df.groupby(['Day'])['sales'].mean().round(2)
print("\nAverage Sales Per Day:")
print(avg_sales_day)
```

Average Sales Per Product

```python
avg_sales_per_product = df.groupby(['product_name'])['sales'].mean().round(2)
print(avg_sales_per_product)
```

```python
avg_sales_per_region = df.groupby(['region'])['sales'].mean().round(2)
print(avg_sales_per_region)
```

Average sales per state

Insights:
- <font size = "2"> Sales are consistently highest in November and December, suggesting a holiday season peak. <font size>
- <font size = "2" > Fridays are the weakest sales days, while Mondays dominate weekly revenue <font size>

Growth Analysis
